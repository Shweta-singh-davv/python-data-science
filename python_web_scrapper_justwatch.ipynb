```python
#importing important modules
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
from random import randint
from time import sleep
import urllib
import seaborn as sns
import matplotlib.pyplot as plt
```

**Task 1:- Web Scrapping**

**Scrapping Movies Data**

```python
#scrapping justwatch website
url= "https://www.justwatch.com/in/movies?release_year_from=2000"
response=requests.get(url)
soup =BeautifulSoup(response.text,'html.parser')
```

```python
print(soup.prettify())
```

**Scrapping Movie URL**

```python
#getting movie links
containers = soup.find_all('a')
movie=[]
for link in containers:
  movie_link= [("https://www.justwatch.com"+link.get('href'))]
  movie = movie+movie_link
#print(movie)
movie=movie[74:]
print(movie)

```

**Getting Movie Title**

```python
name1=[]
movie_1=movie[0:50]


for i in range(0,len(movie_1)):
  new1_url=movie_1[i]
  new1_response=requests.get(str(new1_url))
  new1_soup =BeautifulSoup(new1_response.text,'html.parser')
  name1_container = [new1_soup.find('h1').text]
  name1 = name1+name1_container
print(name1)
```

```python
name2=[]
movie_2=movie[51:100]


for i in range(0,len(movie_2)):
  new2_url=movie_2[i]
  new2_response=requests.get(str(new2_url))
  new2_soup =BeautifulSoup(new2_response.text,'html.parser')
  name2_container = [new2_soup.find('h1').text]
  name2 = name2+name2_container
print(name2)
```

```python
# movie_names list
movie_names=name1 +name2

print(movie_names)
```

**Scrapping Release Year**

```python
# getting year
year1=[]
movie_1=movie[0:50]


for i in range(0,len(movie_1)):
  new1_url=movie_1[i]
  new1_response=requests.get(str(new1_url))
  new1_soup =BeautifulSoup(new1_response.text,'html.parser')
  year1_container = new1_soup.find('span',{"class":"text-muted"})
  if year1_container:
     year1_container = [year1_container.text]
     year1=year1+year1_container
print(year1)







```

```python
year11=[]
for i in year1:
  i=[i.replace('(','').replace(')','').strip()]# replacing special characters and stripping the data off white spaces
  year11=year11+i
print(year11)
```

```python
year2=[]
movie_2=movie[51:100]


for i in range(0,len(movie_2)):
  new2_url=movie_2[i]
  new2_response=requests.get(str(new2_url))
  new2_soup =BeautifulSoup(new2_response.text,'html.parser')
  year2_container = [new2_soup.find('span',{"class":"text-muted"}).text]
  year2 = year2+year2_container
print(year2)
```

```python
year22=[]
for i in year2:
  i=[i.replace('(','').replace(')','').strip()]
  year22=year22+i
print(year22)
```

```python
# final year list
year=year11+year22
print(year)
```

**Scrapping Genre**

```python
# trying genre
movie_url_1=movie[0:50]
genre=[]
for i in range(0,len(movie_url_1)):
  new1_url=movie_url_1[i]
  new1_response=requests.get(str(new1_url))
  new1_soup =BeautifulSoup(new1_response.text,'html.parser')
  genre_1_container=new1_soup.find_all('div',{'class':"detail-infos"})
  for i in genre_1_container:
    if i.find('h3',{'class':"detail-infos__subheading"}).text =='Genres':
      genre.append(i.find('div', {'class':"detail-infos__value"}).text)
print(genre)

```
