```python
# as seen, some columns even though mentioned as object datatype need to be converted to categorical daatatype
# for the analysis to continue . for example, neighbourhood group column is clearly has three distict location names
airbnb['neighbourhood'] = airbnb['neighbourhood'].astype("category")
airbnb['neighbourhood_group'] = airbnb['neighbourhood_group'].astype("category")
airbnb['room_type'] = airbnb['room_type'].astype("category")



```

**DATA WRANGLING**

```python
# Let's begin manipualtion of our data as all the data ids now in correct datatype order.
# firtly, we are going to see rows having null values
null_value= airbnb.isnull().any(axis=1)
airbnb_null = airbnb[null_value]
airbnb_null
```

```python
# most of the nullvalues arein columns last_revies andrevies_per_month.
# let's just keep it as it is and since other columns which have null values
# host name column, lets just keep those rows intact as well
# since host name is not affecting our analysis.



```

```python
# details of top hosts with most reviews in each neighbourhood_group

airbnb.groupby(['host_name','neighbourhood_group'])['number_of_reviews'].max().reset_index()
airbnb.sort_values(by='number_of_reviews', ascending=False).head()
```

```python
# similarly, top hosts with most calculated host listings count
airbnb.groupby(['host_name','neighbourhood_group'])['calculated_host_listings_count'].max().reset_index()
airbnb.sort_values(by='calculated_host_listings_count', ascending=False).head(10)
```

```python
# Let's begin our analysis
# plotting to see most popular locality for the bookings
sns.countplot(x='neighbourhood_group', data= airbnb)

# as can be seen, most popular locality for booking is manhattan and then brooklyn.

```

```python
# booking neighbourhood with respect to price
sns.barplot(x='neighbourhood_group', y='price',data = airbnb)

# with respect to price,top three destinations are manhattan, brooklyn and staten island
```

```python
sns.lineplot(x='price',y='minimum_nights',data=airbnb)
# heavy minimum_nights bookings are below 4000 USD
```

```python
sns.boxplot(x='neighbourhood_group',y='price',data = airbnb)
# most of the price  in every neighbourhood_group is very small but huge number of outliers can be seen in
# manhattan and brooklyn area.
```

```python
sns.scatterplot(x='availability_365',y= 'price', hue='neighbourhood_group',data= airbnb)
# manhattan and brroklyn have the most wide range for availibility with respect to price range
```

```python
sns.countplot(y='room_type',data=airbnb)
# ehtire home/apt is most popular followed by private room
```

```python
# room_type with respect to neighbourhood_group
sns.countplot(x='room_type',hue='neighbourhood_group',data=airbnb)
# manhattan has the most share of entirehome/apt
# brooklyn has the most share in private room types but the margin of entire home/apt and
# private room is close enough to each other in brooklyn
```

```python
# histogram to show price count with respect to each neighbourhoo_group
sns.histplot(data=airbnb,x='price',hue='neighbourhood_group')

```

```python
# histogram for minimum_nights
sns.histplot(data=airbnb,x='minimum_nights')
```
